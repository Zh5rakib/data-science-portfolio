```python
# Project 1: EDA & Visualization — Iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

sns.set(style="whitegrid", context="notebook")

iris = load_iris(as_frame=True)
df = iris.frame.rename(columns={"target": "species_id"})
df["species"] = df["species_id"].map(dict(enumerate(iris.target_names)))

print("Shape:", df.shape)
print(df.head(), "\n")

print("Summary statistics:")
print(df.describe(), "\n")

# Pairplot of features colored by species
sns.pairplot(df, vars=df.columns[:4], hue="species", corner=True, diag_kind="kde")
plt.suptitle("Iris Pairplot", y=1.02)
plt.show()

# Correlation heatmap
corr = df.iloc[:, :4].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation heatmap")
plt.show()

print("Means by species:")
print(df.groupby("species").mean(numeric_only=True))
```
